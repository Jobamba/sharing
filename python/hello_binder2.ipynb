```python
t = np.linspace(0, 6, num=1000)
f1, f2, f3 = 1, 5, 10
y1 = np.cos(2*np.pi*f1*t)
y2 = np.cos(2*np.pi*f2*t) 
y3 = np.cos(2*np.pi*f3*t) 
y_sum = 0.33* (y1 + y2 +y3)
```

```python
plt.figure(figsize=(12, 9))
plt.subplot(411)
plt.plot(t,y1)
plt.subplot(412)
plt.plot(t,y2)
plt.subplot(413)
plt.plot(t,y3)
plt.subplot(414)
plt.plot(t,y_sum)
plt.ylabel('Amplitude')
plt.xlabel('Time')
plt.show()
```
